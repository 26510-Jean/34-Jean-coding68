```python
from  skyfield.api import load

# โหลดข้อมูลดาวเคราะห์และเวลา
planets = load('de421.bsp')  # ไฟล์ข้อมูลตำแหน่งดาว
ts = load.timescale()
t = ts.now()  # เวลาปัจจุบัน

# ดึงข้อมูลของโลกและดาวเสาร์
earth = planets['earth']
saturn = planets['saturn']

# หาตำแหน่งของดาวเสาร์จากโลก
astrometric = earth.at(t).observe(saturn)
ra, dec, distance = astrometric.radec()

# แสดงผล
print(f"เวลาปัจจุบัน (UTC): {t.utc_datetime()}")
print(f"ตำแหน่งของดาวเสาร์:")
print(f" - Right Ascension (RA): {ra}")
print(f" - Declination (Dec): {dec}")
print(f" - Distance from Earth: {distance.km:.2f} km")
```

```python
def kmh_to_ms(speed_kmh):
    speed_ms = speed_kmh * 1000 / 3600
    return speed_ms

# ตัวอย่างการใช้งาน
speed_kmh = float(input("กรอกความเร็ว (km/h): "))
speed_ms = kmh_to_ms(speed_kmh)
print(f"ความเร็ว {speed_kmh} km/h เท่ากับ {speed_ms:.2f} m/s")
```

```python
weight = float(input("กรอกน้ำหนัก (kg): "))
height_cm = float(input("กรอกส่วนสูง (cm): "))

height_m = height_cm / 100
bmi = weight / (height_m **weight)

print("ค่า BMI ของคุณคือ: {bmi:.2f}")
if bmi >= 25:
    print("ผลลัพธ์: ควรควบคุมน้ำหนัก")
else:
    print("ผลลัพธ์: สุขภาพดี")
```
